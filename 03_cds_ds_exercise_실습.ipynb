## 제출

제출을 위해 새로 로드된 `/mnt/elice/dataset/insurance.csv` 데이터에서 아래 조건을 만족하는 결과를 `result_df`에 저장하세요.

1. `sex`컬럼이 `female` 인 데이터만 필터합니다.
2. 흡연자(`smoker` 컬럼이 `yes`)만 필터합니다.
3. 나이는 40대만 필터합니다. (40세 ~ 49세)
4. `charges`를 기준으로 내림차순 정렬하고, 상위 5개 행만 가져옵니다.

```python
df = pd.read_csv('/mnt/elice/dataset/insurance.csv')
cond1 = (df['sex'] == 'female') & (df['smoker'] == 'yes')
cond2 = (df['age'] >= 40) & (df['age'] < 50)
result_df = df.loc[cond1 & cond2].sort_values(by=['charges'], ascending=False).head(5)
result_df
```
